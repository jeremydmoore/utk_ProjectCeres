```python
# importing

# standard libraries
from pathlib import Path

# 3rd party
import pandas as pd
from PIL import Image
```

```python
data_dir_path = Path('./data/TennesseeFarmandHomeSciencePDFs')
```

### create Column B: title and Column D: date year

Form: Periodical title, progress report [#], Month [completely written out] Day, Year


```python
pdf_paths_list = sorted([x for x in data_dir_path.iterdir() if str(x).endswith('.pdf')])
```

```python
# create titles & years
periodical_title = 'Tennessee farm and home science'
quarter = 1
year = 1952

titles_list = []
years_list = []

for scan_number in range(1, 157):  # last scanned issue is 156
    
    identifier = '0012_004267_' + str(scan_number).zfill(6)
    
    if quarter == 1:
        months = 'January - March'
    elif quarter == 2:
        months = 'April - June'
    elif quarter == 3:
        months = 'July - September'
    elif quarter == 4:
        months = 'October - December'
    else:
        print(f'ERROR: cannot have {quarter} quarters!')  # :P
    
    title = f'{periodical_title}, progress report {str(scan_number)}, {months} {year}'
    titles_list.append(title)
    
    years_list.append(year)
    
    quarter += 1
    
    if quarter == 5:
        year += 1
        quarter = 1
    
```

```python
# copy titles list to clipboard for Google Sheets
df = pd.DataFrame(titles_list)
df.to_clipboard(index=False, header=False, excel=True, sep='\t')
```
